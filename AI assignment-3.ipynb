3.Compute outer product of 2 vectors.

```python
import numpy as np
```

```python
p = [[1,0],[0,1]]
q = [[1,2],[3,4]]
print("Given arrays:")
print(p)
print(q)
r = np.outer(p,q)
print("Outer product of two vectors:")
print(r)
```

4.Write a NumPy program to compute the condition number of a given matrix.
Original matrix:

```python
import numpy as np
from numpy import linalg as L
a = np.array([[1,2],[3,4]])
print("Condition number of given matrix: ",L.cond(a))
```

5.Write a NumPy program to compute the sum of the diagonal element of a given array(matrix).

```python
t = np.array([[0,1,2],[3,4,5]])
print("Given Matrix: ")
print(t)
r = np.trace(t)
print("Trace of given matrix:",r)
```
